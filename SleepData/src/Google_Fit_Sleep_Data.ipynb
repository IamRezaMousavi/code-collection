```python
df109 = pd.read_json("Fit\\" + dirsleeplist[109])
df109["startTime"][0], df109["endTime"][0]
```

```python
def createDict(ddff):
    start_time = datetime.datetime.fromisoformat(ddff["startTime"][whichIndex(ddff)].replace("Z", ""))
    end_time = datetime.datetime.fromisoformat(ddff["endTime"][whichIndex(ddff)].replace("Z", ""))

    if datetime.datetime(start_time.year, 3, 21) <= start_time < datetime.datetime(start_time.year, 9, 22):
        start_time += t430
    else:
        start_time += t330
    if datetime.datetime(end_time.year, 3, 21) <= end_time < datetime.datetime(end_time.year, 9, 22):
        end_time += t430
    else:
        end_time += t330

    delta = end_time - start_time
    df7_dict = {
        "startDate": start_time.strftime("%Y-%m-%d"),
        "startTime": start_time.strftime("%H:%M"),
        "endDate": end_time.strftime("%Y-%m-%d"),
        "endTime": end_time.strftime("%H:%M"),
        "sleepTime": str(delta),
        "sleepSeconds": delta.seconds
    }
    return df7_dict

df109_dict = createDict(df109)
df109_dict
```

```python
data = pd.DataFrame(df109_dict, index=[0])
data
```

```python
for i in range(110, 231):
    try:
        dff = pd.read_json("Fit\\" + dirsleeplist[i])
    except:
        dff = pd.DataFrame([pd.read_json("Fit\\" + dirsleeplist[i], typ="series")])
    di = createDict(dff)
    new_data = pd.DataFrame(di, index=[0])
    data = data.append(new_data).reset_index(drop=True)
data
```

```python
data.to_csv("March 21 - Sep 21.csv", index=False)
```
